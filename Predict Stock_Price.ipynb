```python
def treat_outliers_iqr(df):
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
    return df

df = treat_outliers_iqr(df)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Calculate VIF for each independent variable
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif
```

VIF (Variance Inflation Factor) identifies variables that are highly correlated with each other (bad for regression

```python
def stepwise_regression(X, y):
    # Add constant
    X_const = sm.add_constant(X)

    # Initial model
    model = sm.OLS(y, X_const).fit()

    # Columns to drop
    columns_to_drop = []

    while True:
        p_values = model.pvalues[1:] # exclude the constant
        max_p_value = p_values.max()
        if max_p_value > 0.05:
            column_to_drop = p_values.idxmax()
            columns_to_drop.append(column_to_drop)
            X = X.drop(columns=column_to_drop, errors='ignore')  # Corrected line
            X_const = sm.add_constant(X)
            model = sm.OLS(y, X_const).fit()

        else:
            break

    while True:
      vif = pd.DataFrame()
      vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
      vif["features"] = X.columns
      max_vif = vif['VIF Factor'].max()
      if max_vif > 5:
        column_to_drop = vif[vif['VIF Factor'] == max_vif]['features'].values[0]
        columns_to_drop.append(column_to_drop)
        X = X.drop(columns=column_to_drop, errors='ignore')
        X_const = sm.add_constant(X)
        model = sm.OLS(y, X_const).fit()
      else:
        break
    return model, columns_to_drop


X = df.drop(columns=['Stock_Price', 'Date', 'Company', 'Sector','Market_Cap_Log','Assets_Log','Liabilities_Log'], errors='ignore')
y = df['Stock_Price']

final_model, dropped_columns = stepwise_regression(X,y)
print(final_model.summary())
print("Dropped columns:", dropped_columns)
```

# Key Highlights:
R-squared = 0.262 → The model explains 26.2% of the variance in stock prices.

Moderate fit — more variables or interaction terms may improve the model.

All 3 predictors are statistically significant (p < 0.05):

EPS (coef = 5.29): For every unit increase in EPS, the stock price increases by ₹5.29, holding other factors constant. Strong, positive impact.

P/E Ratio (coef = 0.66): For each unit increase in P/E, the stock price increases by ₹0.66. Reflects investor sentiment.

Equity (coef = -1.00): Surprisingly negative, indicating companies with higher equity bases tend to have slightly lower prices, possibly due to overcapitalization or low capital efficiency.

P-values:

All are < 0.05 → highly statistically significant

Durbin-Watson = 2.078 → Residuals are not autocorrelated (good)

# Final Regression Equation:
#** Stock_Price = 21.24 − 1.01 × Equity + 5.29 × EPS + 0.66 × P_E_Ratio**


```python
# Get the fitted values
fitted_values = final_model.fittedvalues

# Get the residuals
residuals = final_model.resid

# Create a residual plot
plt.figure(figsize=(8, 6))
plt.scatter(fitted_values, residuals, alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

The residual plot shows signs of heteroscedasticity, with residuals becoming more dispersed at lower predicted stock prices. This suggests the model’s error variance is not constant, which may affect reliability of coefficient estimates.
